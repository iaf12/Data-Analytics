## EDA

### Rename the columns to an appropriate way

```python
column_dict = {'rank': 'Rank', 'country': 'Country', 'TwoLetterID': 'Two_Letter_ID' ,'unMember':'UN_Membership', 'landAreaKm':'Land_Area_KM' }
df = df.rename(columns=column_dict)
df.head()
```

```python
int_col = df.dtypes[df.dtypes == 'int64'].index
float_col = df.dtypes[df.dtypes == 'float64'].index
obj_col = df.dtypes[df.dtypes == 'object'].index
```

```python
int_col
```

```python
converted_pop = int_col.drop('Rank').drop('2024YoYChange')
converted_pop
```

### Converted population into Millions

```python
for i in converted_pop:
    df[i] = df[i]/1000000
```

```python
df[converted_pop]
```

### Population of Top 20 Countries between 1980 and 2050

```python

for i in converted_pop:
    
    year = df[i].nlargest(20)
    cntry = df.Country[year.index]
    
    plt.figure(figsize=(20,6))
    sns.barplot(cntry, year)
    
    sns.color_palette("husl", 8)
    plt.xlabel('Countries')
    plt.ylabel('Population (M)')
    plt.title('Top 10 Populated countries in {}'.format(i))
```
